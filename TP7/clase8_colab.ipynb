Dado el circuito de la figura encontrar las condiciones que deben cumplir $R_{1}$, $R_{2}$, $C_{1}$ y $C_{2}$ para hacer al sistema no controlable, considerando a $v(t)$ como entrada, y a $v_{1}(t)$ y $v_{2}(t)$ como variables de estado. Hallar la transferencia del sistema para estas condiciones de $R_{1}$, $R_{2}$, $C_{1}$ y $C_{2}$.

Planteamos las ecuaciones y llegamos a:

$$ v(t) = R_1 C_1 \frac{dv_1}{dt} + v_1 $$
$$ v(t) = R_2 C_2 \frac{dv_2}{dt} + v_2 $$

Que nos lleva a la ecuación de estado:

$$ \begin{bmatrix}\dot{v_1} \\ \dot{v_2} \end{bmatrix} = \begin{bmatrix}-\frac{1}{R_1 C_1} & 0 \\ 0 & -\frac{1}{R_2 C_2} \end{bmatrix}\begin{bmatrix}v_1 \\ v_2 \end{bmatrix}+ \begin{bmatrix}\frac{1}{R_1 C_1} \\ \frac{1}{R_2 C_2} \end{bmatrix} v(t) $$

Evaluamos la controlabilidad del sistema y para eso calculamos la matriz:

$$ P = \big[\begin{array}{c|c} B & AB\end{array}\big]$$

$$ P = \begin{bmatrix}\frac{1}{R_1 C_1} & -\frac{1}{(R_1 C_1)^2} \\ \frac{1}{R_2 C_2} & -\frac{1}{(R_2 C_2)^2} \end{bmatrix} $$

Y para que sea no controlable necesitamos que:

$$ r(P) < n = 2$$

En este caso, al ser el sistema tiene única entrada, queda P cuadrada y podemos usar:

$$ det(P) = 0$$

Calculamos:

$$ det(P) = -\frac{1}{R_1 C_1 (R_2 C_2)^2} + \frac{1}{R_2 C_2 (R_1 C_1)^2} = \frac{R_2 C_2 - R_1 C_1}{(R_1 C_1)^2 (R_2 C_2)^2} = 0$$

Se da si:

$$ R_1 C_1 = R_2 C_2 = \tau$$

**Para pensar: ¿Qué significa circuitalmente?**

No nos define una salida el ejercicio. Inventemos una: $ y = v_1 + v_2 $

$$ H(s) = C (sI-A)^{-1} B + D $$

Verificar:

$$ (sI-A)^{-1}= \begin{bmatrix}\frac{1}{s+\frac{1}{R_1 C_1}} & 0 \\ 0 & \frac{1}{s+\frac{1}{R_2 C_2}} \end{bmatrix} $$

$$ H(s) = \begin{bmatrix} 1 & 1 \end{bmatrix} \begin{bmatrix}\frac{1}{s+\frac{1}{R_1 C_1}} & 0 \\ 0 & \frac{1}{s+\frac{1}{R_2 C_2}} \end{bmatrix}\begin{bmatrix}\frac{1}{R_1 C_1} \\ \frac{1}{R_2 C_2} \end{bmatrix} =
\frac{1}{s R_1 C_1 + 1} + \frac{1}{s R_2 C_2 + 1} = \frac{s (R_1 C_1 + R_2 C_2) +2}{(s R_1 C_1 + 1)(s R_2 C_2 + 1)}$$

Para la condición hallada:

$$ H(s) = \frac{s 2 \tau +2}{(s \tau + 1)^2} = \frac{2}{s \tau + 1} $$

Ya que estamos, analizamos la observabilidad del sistema.

**Para pensar: ¿Podemos aplicar la intuición?**

Evaluamos el criterio:

$$ Q = \Big[\begin{array}{c} C \\ \hline CA\end{array}\Big] = \begin{bmatrix} 1 & 1 \\ -\frac{1}{R_1 C_1} & -\frac{1}{R_2 C_2} \end{bmatrix}$$

Notar que como antes, podemos hacer:

$$ det(Q) = -\frac{1}{R_2 C_2 } + \frac{1}{R_1 C_1}$$

También se anula para la condición $ R_1 C_1 = R_2 C_2 = \tau$ y no es observable.

Resolvamos este ejercicio por simulación para entender qué son los modos no controlables y no observables. Dado el sistema:

$$  \begin{eqnarray*}
	\dot{x}(t) &=& \left[\begin{array}{rrr}
		  -7 & -2 & 6 \\
		  2 & -3 & -2 \\
		  -2 & -2 & 1
		\end{array}\right] x(t)  + \left[\begin{array}{rr}
							1 & 1\\
							1 & -1\\
							1 & 0
						   \end{array}\right] u(t) \\
	y(t) &=& \left[\begin{array}{rrr}
			  -1 & -1 & 2 \\
			   1 & 1 & -1
		       \end{array}\right] x(t)
	\end{eqnarray*}
$$

```python
!pip install control
import sympy as symb
import control as ctrl

A = symb.Matrix([[-7, -2, 6], [2, -3, -2], [-2, -2, 1]])
B = symb.Matrix([[1, 1], [1, -1], [1, 0]])
C = symb.Matrix([[-1, -1, 2], [1, 1, -1]])
D = symb.Matrix([[0, 0],[0, 0]])

P = symb.Matrix(ctrl.ctrb(A.tolist(),B.tolist()))
print('\n P =')
symb.pprint(P)

print('El rango de P es', P.rank(), 'entonces no es controlable')

Q = symb.Matrix(ctrl.obsv(A.tolist(),C.tolist()))
print('\n Q =')
symb.pprint(Q)

print('El rango de Q es', Q.rank(), 'entonces no es observable')
```

```python
# No es ni controlable ni observable, llevamos a la forma de Jordan

M, A_hat = A.diagonalize()
print('\n M =')
symb.pprint(M)
print('\n A_hat =')
symb.pprint(A_hat)

# Hallamos el resto de las matrices:

B_hat = M.inv()*B
print('\n B_hat =')
symb.pprint(B_hat)
C_hat = C*M
print('\n C_hat =')
symb.pprint(C_hat)
```
